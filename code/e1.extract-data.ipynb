```python
import spacy
from spacy import displacy

import pandas as pd

# use GPU if available
spacy.prefer_gpu()
```

```python
# load the trained model
nlp = spacy.load('en_udrp_extractor')

# read the file containing the dispute information collected from the providers
df = pd.read_json('../data/disputes-content.jsonl.gz', lines = True)
```

```python
# take a random element from the proceedings and process it using the model
doc = nlp(df[df.language == "en"].sample(n = 1).text.values[0])

# show the model's annotations
displacy.render(doc, style = 'ent', jupyter = True)
```
